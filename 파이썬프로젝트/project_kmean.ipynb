```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from random import randint
import matplotlib
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
steam = pd.read_csv("gamelist2.csv", index_col='game_num', encoding='cp949')
steam
```

```python
game = pd.DataFrame(steam)
game
```

```python
### 원핫인코딩
cata =  {"genre","thema",}
for feature_name in cata:
    one_hot = pd.get_dummies( game[feature_name], prefix = feature_name)# 원 핫 인코딩
    game.drop(feature_name, axis = 1, inplace =True) # 기존 문자형태 컬럼 삭제
    game = pd.concat([game, one_hot], axis = 1)
game
```

```python
game.columns
```

```python
x= game.drop('playtime',axis=1)
x= x.drop('image',axis=1)
x_data= x.drop('game_name',axis=1)
x_target=game.loc[:,['game_name']]
print(x_data)
print(x_target)
```

from sklearn.preprocessing import StandardScaler
x_result = StandardScaler().fit_transform(x)
features = ['graphic', 'atmos', 'multi', 'price', 'lv', 'playtime', 'thema_19금', 'thema_sf', 'thema_공포', 'thema_미스터리',

x_result

```python
kmeans = KMeans(n_clusters=16, random_state=0).fit(x_data)
kmeans.labels_
```

```python
y=pd.DataFrame(kmeans.labels_)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_data,
                                                   y,
                                                   test_size=0.2,
                                                   random_state=10)
```

```python
model = RandomForestClassifier(random_state=5)

params = {'n_estimators':[60,100,151,200],
         'max_depth':[2,3,5,7],
         'max_leaf_nodes':[4,6,8],
         'min_samples_split':[5,50,100]}

grid = GridSearchCV(model,
                   params,
                   cv=10,
                   scoring='accuracy')

grid.fit(X_train,y_train)
print(grid.best_params_)
print(grid.best_score_)
```

```python
model_test = RandomForestClassifier(random_state=5, max_depth=7, max_leaf_nodes=8, min_samples_split=5)

result2 = model_test.fit(X_train,y_train)

train_pred2 = model_test.predict(X_train)
test_pred2 = model_test.predict(X_test)

print('학습 데이터 셋의 정확도 = ', accuracy_score(y_train, train_pred2))
print('테스트 데이터 셋의 정확도 = ', accuracy_score(y_test, test_pred2))
```

```python
model2 = XGBClassifier()

# 튜닝할 매개변수 설정
params = {'learning_rate':[0.05,0.1,0.3,0.5],
         'max_depth':[3,5,7,11],
         'n_estimators':[60,100,150,300]}

# 쌈싸기
grid2 = GridSearchCV(model2,
                    params,
                    cv=10,
                    scoring='accuracy')
grid2.fit(X_train,y_train)

print(grid2.best_params_)
print(grid2.best_score_)
```

```python
model_best = XGBClassifier(learning_rate=0.3,
                          max_depth=5,
                          n_estimators=150)

result1 = model_best.fit(X_train,y_train)

train_pred = model_best.predict(X_train)
test_pred = model_best.predict(X_test)

# 예측 결과 확인(시험지 확인)
print('학습 데이터를 이용한 예측의 결과\n', train_pred)
print('테스트 데이터를 이용한 예측의 결과\n', test_pred)
```

```python
print('학습 데이터 셋의 정확도 = ', accuracy_score(y_train, train_pred))
print('테스트 데이터 셋의 정확도 = ', accuracy_score(y_test, test_pred))
```
