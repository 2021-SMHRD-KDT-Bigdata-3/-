# BeautifulSoup 사용하기
- 가져온 페이지 정보에서 내가 원하는 정보들만 추출하게 도와주는 라이브러리

```python
from bs4 import BeautifulSoup as bs #bs4폴더안의 beautifulSoup 라는 라이브러리 가져와줘~
```

```python
#bs(어떤 데이터를 가지고 가공할건지, 어떻게 가공할건지 -파싱방법를 가지고 가공할건지, 어떻게 가공할건지 -파싱방법)
naver = bs(res.text,"lxml") # "lxml" ->가공하는 데이터의 방법중 하나.html.parser도 있음.
```

```python
#find_all >> 데이터가 여러개가 넘어온다!! >> 리턴타입 리스트!
result = naver.find_all("a",class_ = "nav") # 네이버 블로그에 마우스 대고 커서 오른쪽 클릭하고 검사 해바. .find_all("mark",class_ = "nav") class같은 기능들은 _(언더바)필요.
# 현재 a class는 리스트형태다. []로 감싸져 있으므로
```

```python
result[2] #변수 result는 리스트형태므로 인덱싱 가능 # 현재 블로그 인덱스 번호 2번으로 블로그만 가져와봤다.
```

```python
result[2].text #블로그안에 태그 다 뺴고 글자만 보고싶어!
```

```python
result[0].text
```

```python
result[1].text
```

# 멜론 TOP 100 차트 가지고 오기
-노래제목 가져오기

```python
head = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36 Edg/89.0.774.63"}
```

```python
read = req.get("https://www.melon.com/chart/index.htm",headers = head)
```

```python
naver2 = bs(read.text,"lxml")
```

```python
result3 = naver2.find_all("div",class_ = "ellipsis rank01")
```

```python
result4 = naver2.find_all("span",class_ = "checkEllipsis")
```

```python
for i in result3:
    print(i.text.strip()) #.strip() 필요없는 개행문자같은 문자들 제거할 때 사용
```

```python
for i in result4:
    print(i.text)
```

```python
## 가수명 가지고 오기!
naver2.select("div.ellipsis.rank02") # .select()는 가져올때 마크이름,하고 id면 # class면 .찍고 바로 이름적어준다
# ellipsis rank02 가 공백인 이유 = ellipsis하나, rank02하나로 나뉘어져서 공백으로 뜨는거임.그래서 공백을 없애주고 .(class)를 해주면 나오게된다.
```

```python
result4 = naver2.select("span.checkEllipsis") # "div.ellipsis.rank02"안에 "span.checkEllipsis"에도 가수이름이 있기 때문에 2번씩 뜬거다! 
#span 위에 있는 a는 class가 없으므로 span 안의 class를 가져온다
```

```python
for i in result4:
    print(i.text.strip())
```

```python
#데이터프레임() 즉 데이터 출력
```

```python
# 순수한 글자 정보만 담아줄 리스트
rank_list = []
song_list = []
artist_list = []
```

```python
for i in range(len(result3)): #range(result3의 길이값)의 길이 만큼 돌거에요~
    rank_list.append(i+1)
    song_list.append(result3[i].text.strip()) # append(result)
    artist_list.append(result4[i].text.strip())
    
```

```python
music_dic = {'rank': rank_list, 'song': song_list, 'artist': artist_list}
```

```python
music_dic
```

```python
import pandas as pd
```

```python
music_df = pd.DataFrame(music_dic)
```

```python
music_df.set_index('rank', inplace = True) #.set_index(바꿀이름) 인덱스 제목바꾸는법 
music_df #set_index('값', inplace = True)는 저장하는것
```

```python
music_df.to_csv('melon_song2.csv', encoding = "") #.to_csv() ()안으로 보내겠다
```
